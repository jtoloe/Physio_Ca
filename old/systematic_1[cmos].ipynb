```python
def decay(time,top,bottom,rate):
#     top,bottom,rate = p
    return bottom+(top-bottom)*np.exp(-time*rate)
def Hill(conc,p):
    if len(p)==4:
        top,bottom,ic50,h = p
    elif len(p)==3:
        top,bottom,ic50 = p
        h = 1
    else:
        raise ValueError("There should be three of four parameters")
    return 1./(1+conc/ic50)**h*(top-bottom)+bottom

def HillLog(log10conc,p):
    if len(p)==4:
        top,bottom,logIC50,h = p
    elif len(p)==3:
        top,bottom,logIC50 = p
        h = 1
    else:
        raise ValueError("There should be three of four parameters")
    return 1./(1+10**(log10conc-logIC50))**h*(top-bottom)+bottom
```
