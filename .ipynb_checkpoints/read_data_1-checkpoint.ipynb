```python
bin = 20
spins = []
for k in xrange(len(ts)):
    x = ts[k]
    N  = 2
    Wn = 0.005
    B, A = signal.butter(N, Wn, output='ba')
    tsf_low = signal.filtfilt(B,A, x)
    N  = 2
    Wn = 0.2
    B, A = signal.butter(N, Wn, output='ba')
    tsf_high = signal.filtfilt(B,A, x)
    y = tsf_high-tsf_low
    y_scaled = (y-min(y))/(max(y)-min(y))
    n = len(y_scaled)
    s = np.empty(n//bin)
    s.fill(-1)
    y_cutoff = np.mean(y_scaled[:4000])
    y_std = np.std(y_scaled[:4000])
    for i in xrange(0,n-bin,bin):
        if np.mean(y_scaled[i:i+bin]) > y_cutoff+3*y_std:
            s[i//bin] = 1
    spins.append(s)
    
# firing rate
fr = [(sum(spins[k])/len(spins[k])+1)/2/bin for k in xrange(len(spins))]
inz = np.where(np.array(fr) > 0.0)[0]
nzfr = [fr[k] for k in inz]
s = [spins[i] for i in inz]
numspikes = [sum(spins[k]+1) for k in xrange(len(spins))]
idx = np.where(np.array(numspikes) >= 15.0)[0]
s = [spins[i] for i in idx]
pos = [pos[i] for i in idx]
```

```python
plt.plot(spins[110])
```

```python
cm = np.cov(s)
cv = []
for i in xrange(len(cm)):
	for j in xrange(i+1,len(cm)):
		cv.append(cm[i,j])
```

```python
_,_,_ = plt.hist(cv, bins=50,normed=1,color='white')
```

```python
d = []
crds = []
n = len(pos)
for i in xrange(n):
	for j in xrange(i+1,n):
		d.append(np.sqrt((pos[i][0]-pos[j][0])**2+(pos[i][1]-pos[j][1])**2))
		crds.append([i,j])
```

```python
dd = 10.0
cav = []
dav = []
cstd = []
counts = []
corr = cv
for c in np.linspace(min(d),max(d)-dd,int((max(d)-dd-min(d))/dd)):
	idx = np.where((np.array(d) < c+dd) & (np.array(d) > c))[0]
	counts.append(len(idx))
	cav.append(np.mean([corr[i] for i in idx]))
	dav.append(np.mean([d[i] for i in idx]))
	cstd.append(np.std([corr[i] for i in idx]))
```

```python
pixel_length = 317.32/256
plt.plot(pixel_length*np.array(d[::10]),corr[::10],ls='',marker='.',mfc='none',lw=1.5,mew=1.5,ms=8,color='lightgray')
plt.plot(pixel_length*np.array(dav),cav,ls='--',marker='+',lw=1.0,color='black',ms=12)
#plt.plot(dav,var/np.max(var),ls='--',marker='.',lw=1.0,color='black',ms=11)
plt.xlabel('cell-cell distance, um',fontsize=16)
plt.ylabel('correlations',fontsize=16)
```
