### Activation time

so, when does a particular roi activate?

```python
roi = 341
# events.roi.value_counts().index[0]
```

```python
# simple case
ax = regions.plot_trace(roi)
ax.axvline( events.query(f"roi=={roi}").t0.min(), color="C1", ls="--")
# ax.legend()
```

```python
regions.protocol
```

```python
events["tend"] = events['t0']+events["halfwidth"]
```

```python
# but what is we are interested for a min t0 but only in the low Ca leg of the exp?
twindow = (2119.0, 3838.)
ax = regions.plot_trace(roi)
ax.axvline( events.query(f"roi=={roi} and t0>{twindow[0]} and tend<{twindow[1]}").t0.min(), color="C1", ls="--", label="advanced")
ax.legend()
```

```python
# so far so good, but what if, in addition to above constraint, we only care if the event was long enough, say 20s?
twindow = (2119.0, 3838.)
hw_threshold = 20
ax = regions.plot_trace(roi)
ax.axvline( events.query(f"roi=={roi} and t0>{twindow[0]} and tend<{twindow[1]} and halfwidth>{hw_threshold}").t0.min(), color="C1", ls="--", label="advanced")
ax.legend()
```

I hope you get the point.

Now to get those values for all rois in a single command

```python
times = np.array([events.query(f"roi=={roi}").t0.min() for roi in rois])
plt.hist(times,100);
plt.title("time of first event");
```

This can even be used as a thing to filter on: if there is no event in a roi within, say, first 1000s, we might even discard this roi alltoghether.

```python
accept_rois = rois[times<1000]
```

```python
# watch out: this is irreversible:
events = events[events.roi.isin(accept_rois)]
```

```python
# to go back to an example from above:
# we are interested only in the low Ca leg of the exp, and only those events which last more than 20s

times = np.array([events.query(f"roi=={roi} and halfwidth>{hw_threshold} and t0>{twindow[0]} and tend<{twindow[1]}").t0.min() for roi in accept_rois])
plt.hist(times,100);
```

```python
print(f"there are {np.isnan(times).sum()} rois which have no event that satisfies criteria.")
```

```python
# Now, this is just a dataset that you can plot a boxplot of
plt.boxplot(times[np.isfinite(times)])
```
