```python
from scipy.spatial import Voronoi, voronoi_plot_2d, qhull
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# points = np.random.rand(10).reshape(-1,2)
points = [[0,0],[0,1],[1,1],[.5,.5],[1,0]]
```

```python
vor = Voronoi(points, )
```

```python
plt.figure(figsize=(6,6))
ax = plt.subplot(111)
voronoi_plot_2d(vor,ax=ax)
for i,p in enumerate(vor.points):
    ax.text(*p,i)
for i,p in enumerate(vor.vertices):
    ax.text(*p,i)

for rp in vor.ridge_points:
    plt.plot(*vor.points[rp].T, color="g",lw=.3)
    break
ax.set_ylim(-1,2)
ax.set_xlim(-1,2)
ax.set_aspect("equal")
```
